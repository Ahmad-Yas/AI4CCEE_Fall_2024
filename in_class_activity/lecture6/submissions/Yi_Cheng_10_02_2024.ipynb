```python
# Import necessary libraries
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
from pyspark.sql.types import IntegerType, DoubleType, TimestampType
import matplotlib.pyplot as plt
import seaborn as sns
import logging
import pandas as pd

# Initialize logging
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

# Initialize the Spark session
spark = SparkSession.builder \
    .appName("Speed Analysis") \
    .getOrCreate()

# Load CSV File Using PySpark
file_path = "/content/ID_31_data.csv"  # Change this to your CSV file path
df_spark = spark.read.csv(file_path, header=True, inferSchema=True)

# Ensure 'speed' is of type Double and 'ID' is of type Integer
df_spark = df_spark.withColumn("speed", df_spark["speed"].cast(DoubleType())) \
                   .withColumn("ID", df_spark["ID"].cast(IntegerType()))

# Convert 'capturedtimestamp' to Spark TimestampType
df_spark = df_spark.withColumn("capturedtimestamp", df_spark["capturedtimestamp"].cast(TimestampType()))

# Extract the hour and day of the week for further analysis
df_spark = df_spark.withColumn("hour", F.hour("capturedtimestamp"))
df_spark = df_spark.withColumn("day_of_week", F.dayofweek("capturedtimestamp"))

# Calculate the required statistics using PySpark functions
speed_limit_value = 60  # Adjust this to your actual speed limit

# Total count of rows in the DataFrame
total_count = df_spark.count()

# Calculate fractions
fraction_5_over_limit = df_spark.filter(df_spark["speed"] > (speed_limit_value + 5)).count() / total_count
fraction_10_over_limit = df_spark.filter(df_spark["speed"] > (speed_limit_value + 10)).count() / total_count
fraction_20_over_limit = df_spark.filter(df_spark["speed"] > (speed_limit_value + 20)).count() / total_count

# Log the computed statistics
logger.info(f"Fraction 5 Over Limit: {fraction_5_over_limit}")
logger.info(f"Fraction 10 Over Limit: {fraction_10_over_limit}")
logger.info(f"Fraction 20 Over Limit: {fraction_20_over_limit}")

# Calculate 50th percentile (median) speed per journey
median_speed_per_journey = df_spark.groupBy("ID").agg(F.expr("percentile_approx(speed, 0.5)").alias("median_speed"))

# Calculate fractions of median speed over limits
median_fraction_5_over_limit = median_speed_per_journey.filter(median_speed_per_journey["median_speed"] > (speed_limit_value + 5)).count() / median_speed_per_journey.count()
median_fraction_10_over_limit = median_speed_per_journey.filter(median_speed_per_journey["median_speed"] > (speed_limit_value + 10)).count() / median_speed_per_journey.count()
median_fraction_20_over_limit = median_speed_per_journey.filter(median_speed_per_journey["median_speed"] > (speed_limit_value + 20)).count() / median_speed_per_journey.count()

# Log the median statistics
logger.info(f"Median Fraction 5 Over Limit: {median_fraction_5_over_limit}")
logger.info(f"Median Fraction 10 Over Limit: {median_fraction_10_over_limit}")
logger.info(f"Median Fraction 20 Over Limit: {median_fraction_20_over_limit}")

# Print all statistics
print(f"Fraction 5 Over Limit: {fraction_5_over_limit}")
print(f"Fraction 10 Over Limit: {fraction_10_over_limit}")
print(f"Fraction 20 Over Limit: {fraction_20_over_limit}")
print(f"Median Fraction 5 Over Limit: {median_fraction_5_over_limit}")
print(f"Median Fraction 10 Over Limit: {median_fraction_10_over_limit}")
print(f"Median Fraction 20 Over Limit: {median_fraction_20_over_limit}")

# Convert to Pandas DataFrame for plotting
df_pandas = df_spark.toPandas()

# Plot Boxplots to show Speed by Time of Day (Weekdays)
plt.figure(figsize=(12, 6))
sns.boxplot(x='hour', y='speed', data=df_pandas[df_pandas['day_of_week'] <= 5])  # Weekdays
plt.title('Speed by Time of Day (Weekdays)')
plt.xlabel('Hour of Day')
plt.ylabel('Speed')
plt.show()

# Plot Boxplots to show Speed by Time of Day (Weekends)
plt.figure(figsize=(12, 6))
sns.boxplot(x='hour', y='speed', data=df_pandas[df_pandas['day_of_week'] > 5])  # Weekends
plt.title('Speed by Time of Day (Weekends)')
plt.xlabel('Hour of Day')
plt.ylabel('Speed')
plt.show()

# Stop the Spark session
spark.stop()
```

```python
# Step 1: Install PySpark
!pip install pyspark

# Step 2: Create Spark Session
from pyspark.sql import SparkSession

# Create Spark session
spark = SparkSession.builder \
    .appName("PySpark JSON Parsing") \
    .getOrCreate()

# Step 3: Upload a File into RDD
from google.colab import files
import json

# Upload a file
uploaded = files.upload()

# Specify the name of the uploaded file
file_name = list(uploaded.keys())[0]

# Create RDD from the uploaded JSON file
rdd = spark.sparkContext.textFile(file_name)

# Step 4: Inspect First Few Rows
print("First few rows of RDD:")
for row in rdd.take(5):
    print(row)

# Step 5: Parse JSON Strings into Python Dictionaries
json_rdd = rdd.map(lambda x: json.loads(x))

# Inspect the first few parsed rows
print("Parsed JSON Rows:")
for row in json_rdd.take(5):
    print(row)
```

```python
# Step 6: Extract Key Information (latitude, longitude, speed)
# Assuming the JSON structure is like: {"latitude": value, "longitude": value, "speed": value}

rdd_extracted = json_rdd.map(lambda x: (x['location']['latitude'], x['location']['longitude'], x['metrics']['speed']))
rdd_extracted.take(5)

```

```python


# Step 3: Filter Data Points with Speed > 30 mph
# Check if speed is not None before comparing
filtered_rdd = rdd_extracted.filter(lambda x: x[2] is not None and x[2] > 30)  # Assuming speed is the 3rd element in the tuple

# Step 4: Count the Number of Data Points with Speed > 30 mph
count_speed_over_30 = filtered_rdd.count()
print(f"Number of Data Points with Speed > 30 mph: {count_speed_over_30}")

# Step 5: Collect and Display Filtered Data Points
filtered_data = filtered_rdd.collect()
print("Filtered Data Points (latitude, longitude, speed):")
for row in filtered_data[:5]:  # Display the first 5 for brevity
    print(row)
```

```python
df = filtered_rdd.toDF(["latitude", "longitude", "speed"])
import matplotlib.pyplot as plt


plt.scatter(df.select('latitude').collect(), df.select('longitude').collect(), c=df.select('speed').collect())
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.title('Scatter Plot of Latitude vs Longitude')
plt.colorbar(label='Speed')
plt.show()
```
