```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read data to create dataframe
df = pd.read_csv('Wavetronix.csv')
```

```python
# Inspect dataframe
print(df.head())
```

```python
print(df.describe())
```

```python
# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['Time'], df['WAVE-SW-Speed'], color='blue', alpha=0.6)

# Add labels and title
plt.xlabel('Time')
plt.ylabel('WAVE-SW-Speed')
plt.title('Time vs WAVE-SW-Speed')

plt.show()
```

```python
# Filter the DataFrame for Day 1
df_day1 = df[df['day'] == 1]

# Define the threshold
threshold = 40

# Filter the DataFrame for Day 1 where WAVE-SW-Speed is greater than the threshold
df_speed = df[(df['day'] == 1) & (df['WAVE-SW-Speed'] > threshold)]
```

```python
# Count the number of times speed drops below threshold
count = (df_speed['WAVE-SW-Speed'] < threshold).sum()

# Print the count
print(f'The speed drops below {threshold} a total of {count} times.')
```

```python
print(df_day1.head())
```

```python
# Create a scatter plot
plt.figure(figsize=(8, 6))

# Plot the WAVE-SW-Speed data for df_day1 (in blue)
plt.scatter(df_day1['Time'], df_day1['WAVE-SW-Speed'], color='red', alpha=0.6, label='df_day1 WAVE-SW-Speed')

# Plot the WAVE-SW-Speed data for df_speed (in red)
plt.scatter(df_speed['Time'], df_speed['WAVE-SW-Speed'], color='blue', alpha=0.6, label='df_speed WAVE-SW-Speed')

# Add labels and title
plt.xlabel('Time')
plt.ylabel('WAVE-SW-Speed')
plt.title('Time vs WAVE-SW-Speed')

# Add a legend to distinguish between the two datasets
plt.legend()

# Display the plot
plt.show()
```

```python
df_day1['color'] = np.where(df_day1['WAVE-SW-Speed'] < 40, 'red', 'blue')
plt.scatter(df_day1['Time'], df_day1['WAVE-SW-Speed'], c=df_day1['color'])

# Add labels and title
plt.xlabel('Time')
plt.ylabel('WAVE-SW-Speed')
plt.title('Time vs WAVE-SW-Speed')

# Display the plot
plt.show()
```

```python
# Calculate rolling average with a window of 5
df_speed['Rolling_Average'] = df_speed['WAVE-SW-Speed'].rolling(window=5).mean()
```

```python
plt.scatter(df_speed['Time'], df_speed['Rolling_Average'], color='blue', alpha=0.6)

# Add labels and title
plt.xlabel('Time')
plt.ylabel('Rolling Average')
plt.title('Time vs Rolling Average')

# Display the plot
plt.show()
```

```python
# Remove missing data from the 'WAVE-SW-Speed' column in df_speed
df_speed_cleaned = df_speed.dropna(subset=['WAVE-SW-Speed'])

# Display the cleaned DataFrame
df_speed_cleaned.head()
```

```python
# Remove duplicate rows from df_speed_cleaned
df_speed_cleaned = df_speed_cleaned.drop_duplicates()

# Display the cleaned DataFrame
df_speed_cleaned.head()
```

```python
# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df_speed_cleaned['Time'], df_speed_cleaned['WAVE-SW-Speed'], color='blue', alpha=0.6)

# Add labels and title
plt.xlabel('Time')
plt.ylabel('WAVE-SW-Speed')
plt.title('Time vs WAVE-SW-Speed')

plt.show()
```

```python
# Apply FFT to the 'WAVE-SW-Speed' column
wave_sw_speed = df_speed_cleaned['WAVE-SW-Speed'].values  # Convert to numpy array
fft_result = np.fft.fft(wave_sw_speed)

# Get the frequencies corresponding to the FFT result
n = len(wave_sw_speed)  # Length of the data
frequencies = np.fft.fftfreq(n)

# Plot the FFT results
plt.figure(figsize=(10, 6))

# Plot the magnitude of the FFT result (absolute value)
plt.plot(frequencies, np.abs(fft_result), color='blue')

# Add labels and title
plt.xlabel('Frequency')
plt.ylabel('Magnitude')
plt.title('FFT of WAVE-SW-Speed')

# Display the plot
plt.show()
```

```python
#!/usr/bin/env python

"""Plot the  set of 1D demo signals available in `pywt.data.demo_signal`."""

import matplotlib.pyplot as plt
import numpy as np

# Install pywt using pip
!pip install PyWavelets

# Import the pywt module after installation
import pywt

# use 'list' to get a list of all available 1d demo signals
signals = pywt.data.demo_signal('list')

subplots_per_fig = 5
signal_length = 1024
i_fig = 0
n_figures = int(np.ceil(len(signals)/subplots_per_fig))
for i_fig in range(n_figures):
    # Select a subset of functions for the current plot
    func_subset = signals[
        i_fig * subplots_per_fig:(i_fig + 1) * subplots_per_fig]

    # create a figure to hold this subset of the functions
    fig, axes = plt.subplots(subplots_per_fig, 1)
    axes = axes.ravel()
    for n, signal in enumerate(func_subset):
        if signal in ['Gabor', 'sineoneoverx']:
            # user cannot specify a length for these two
            x = pywt.data.demo_signal(signal)
        else:
            x = pywt.data.demo_signal(signal, signal_length)
        ax = axes[n]
        ax.plot(x.real)
        if signal == 'Gabor':
            # The Gabor signal is complex-valued
            ax.plot(x.imag)
            ax.legend(['Gabor (Re)', 'Gabor (Im)'], loc='upper left')
        else:
            ax.legend([signal, ], loc='upper left')
    # omit axes for any unused subplots
    for n in range(n + 1, len(axes)):
        axes[n].set_axis_off()
plt.show()
```

```python
#!/usr/bin/env python

import matplotlib.pyplot as plt
import numpy as np

import pywt

time, sst = pywt.data.nino()
dt = time[1] - time[0]

# Taken from http://nicolasfauchereau.github.io/climatecode/posts/wavelet-analysis-in-python/
wavelet = 'cmor1.5-1.0'
scales = np.arange(1, 128)

[cfs, frequencies] = pywt.cwt(sst, scales, wavelet, dt)
power = (abs(cfs)) ** 2

period = 1. / frequencies
levels = [0.0625, 0.125, 0.25, 0.5, 1, 2, 4, 8]
f, ax = plt.subplots(figsize=(15, 10))
ax.contourf(time, np.log2(period), np.log2(power), np.log2(levels),
            extend='both')

ax.set_title(f'Nino1+2 Wavelet Power Spectrum ({wavelet})')
ax.set_ylabel('Period (years)')
Yticks = 2 ** np.arange(np.ceil(np.log2(period.min())),
                        np.ceil(np.log2(period.max())))
ax.set_yticks(np.log2(Yticks))
ax.set_yticklabels(Yticks)
ax.invert_yaxis()
ylim = ax.get_ylim()
ax.set_ylim(ylim[0], -1)

plt.show()
```

```python
from PIL import Image
# Load the image using Pillow
image = Image.open('2018-06-30-19-05.jpg')  # image path
```

```python
# Display the image using Matplotlib
plt.imshow(image)
plt.axis('off')  # Turn off the axis labels
plt.title('Original Image')  # Add a title if you want
plt.show()
```

```python
# Convert the image to grayscale
img_gray = image.convert('L')  # 'L' mode means grayscale

# Convert the grayscale image to a NumPy array
img_array = np.array(img_gray)

# Perform 2D Discrete Wavelet Transform using the 'haar' wavelet
coeffs2 = pywt.dwt2(img_array, 'haar')

# Extract the approximation and detail coefficients
cA, (cH, cV, cD) = coeffs2

# Plot the decomposed components
plt.figure(figsize=(12, 8))

# Plot the approximation coefficients (cA)
plt.subplot(2, 2, 1)
plt.imshow(cA, cmap='gray')
plt.title('Approximation (cA)')

# Plot the horizontal detail coefficients (cH)
plt.subplot(2, 2, 2)
plt.imshow(cH, cmap='gray')
plt.title('Horizontal Detail (cH)')

# Plot the vertical detail coefficients (cV)
plt.subplot(2, 2, 3)
plt.imshow(cV, cmap='gray')
plt.title('Vertical Detail (cV)')

# Plot the diagonal detail coefficients (cD)
plt.subplot(2, 2, 4)
plt.imshow(cD, cmap='gray')
plt.title('Diagonal Detail (cD)')

# Display the plots
plt.tight_layout()
plt.show()
```
