```python
# prompt: Implement a 5-minute rolling average to smooth the sensor data for the first 1253 rows

# Assuming your DataFrame has a 'WAVE-SW-Speed' column and 'Time' column
rolling_avg_speed = df['WAVE-SW-Speed'][0:1253].rolling(window=5).mean()

# Create a plot to visualize the original data and the rolling average
plt.figure(figsize=(12, 6))
plt.plot(df['Time'][0:1253], df['WAVE-SW-Speed'][0:1253], label='Original Speed')
plt.plot(df['Time'][0:1253], rolling_avg_speed, label='5-Minute Rolling Average')
plt.xlabel('Time')
plt.ylabel('Speed')
plt.title('Speed vs. Time with 5-Minute Rolling Average')
plt.legend()
plt.show()
```

```python
# prompt: Apply the Fast Fourier Transform (FFT) to the 5-minute rolling average to smooth the sensor data for the first 1253 rows data

# Assuming 'rolling_avg_speed' is your 5-minute rolling average data

# Apply FFT
from scipy.fft import fft, fftfreq

yf = fft(rolling_avg_speed)
xf = fftfreq(len(rolling_avg_speed), 1)  # Assuming a sampling rate of 1 (adjust if needed)

# You can plot the FFT results if you want to analyze the frequency components
# plt.plot(xf, np.abs(yf))
# plt.xlabel('Frequency')
# plt.ylabel('Magnitude')
# plt.title('FFT of Rolling Average Speed')
# plt.show()

# Apply a low-pass filter (e.g., keep only frequencies below a certain threshold)
# This will smooth the data by removing high-frequency components
low_pass_cutoff = 0.1  # Adjust the cutoff frequency as needed
yf_filtered = yf.copy()
yf_filtered[(np.abs(xf) > low_pass_cutoff)] = 0

# Inverse FFT to get the smoothed data
smoothed_speed = np.fft.ifft(yf_filtered)

# Plot the original and smoothed data
plt.figure(figsize=(12, 6))
plt.plot(df['Time'][0:1253], rolling_avg_speed, label='5-Minute Rolling Average')
plt.plot(df['Time'][0:1253], smoothed_speed, label='Smoothed Speed (FFT)')
plt.xlabel('Time')
plt.ylabel('Speed')
plt.title('Speed vs. Time with FFT Smoothing')
plt.legend()
plt.show()
```

```python
#!/usr/bin/env python

! pip install pywavelets
"""Plot the  set of 1D demo signals available in `pywt.data.demo_signal`."""

import matplotlib.pyplot as plt
import numpy as np

import pywt

# use 'list' to get a list of all available 1d demo signals
signals = pywt.data.demo_signal('list')

subplots_per_fig = 5
signal_length = 1024
i_fig = 0
n_figures = int(np.ceil(len(signals)/subplots_per_fig))
for i_fig in range(n_figures):
    # Select a subset of functions for the current plot
    func_subset = signals[
        i_fig * subplots_per_fig:(i_fig + 1) * subplots_per_fig]

    # create a figure to hold this subset of the functions
    fig, axes = plt.subplots(subplots_per_fig, 1)
    axes = axes.ravel()
    for n, signal in enumerate(func_subset):
        if signal in ['Gabor', 'sineoneoverx']:
            # user cannot specify a length for these two
            x = pywt.data.demo_signal(signal)
        else:
            x = pywt.data.demo_signal(signal, signal_length)
        ax = axes[n]
        ax.plot(x.real)
        if signal == 'Gabor':
            # The Gabor signal is complex-valued
            ax.plot(x.imag)
            ax.legend(['Gabor (Re)', 'Gabor (Im)'], loc='upper left')
        else:
            ax.legend([signal, ], loc='upper left')
    # omit axes for any unused subplots
    for n in range(n + 1, len(axes)):
        axes[n].set_axis_off()
plt.show()
```

```python
#!/usr/bin/env python

import matplotlib.pyplot as plt
import numpy as np

import pywt

time, sst = pywt.data.nino()
dt = time[1] - time[0]

# Taken from http://nicolasfauchereau.github.io/climatecode/posts/wavelet-analysis-in-python/
wavelet = 'cmor1.5-1.0'
scales = np.arange(1, 128)

[cfs, frequencies] = pywt.cwt(sst, scales, wavelet, dt)
power = (abs(cfs)) ** 2

period = 1. / frequencies
levels = [0.0625, 0.125, 0.25, 0.5, 1, 2, 4, 8]
f, ax = plt.subplots(figsize=(15, 10))
ax.contourf(time, np.log2(period), np.log2(power), np.log2(levels),
            extend='both')

ax.set_title(f'Nino1+2 Wavelet Power Spectrum ({wavelet})')
ax.set_ylabel('Period (years)')
Yticks = 2 ** np.arange(np.ceil(np.log2(period.min())),
                        np.ceil(np.log2(period.max())))
ax.set_yticks(np.log2(Yticks))
ax.set_yticklabels(Yticks)
ax.invert_yaxis()
ylim = ax.get_ylim()
ax.set_ylim(ylim[0], -1)

plt.show()
```

```python
# prompt: Perform wavelet transform: use dwt2 to decompose the Image('/content/2018-06-30-19-05.jpg') and use haar wavelet

import pywt
import numpy as np
from PIL import Image

# Replace 'image_path' with the actual path to your image
image_path = '/content/2018-06-30-19-05.jpg'

# Open the image using PIL
image = Image.open(image_path).convert('L')  # Convert to grayscale

# Convert the image to a NumPy array
image_array = np.array(image)

# Perform 2D discrete wavelet transform using Haar wavelet
coeffs = pywt.dwt2(image_array, 'haar')

# coeffs is a tuple containing the approximation coefficients (cA)
# and the horizontal (cH), vertical (cV), and diagonal (cD) detail coefficients.
cA, (cH, cV, cD) = coeffs

# You can now work with the wavelet coefficients (cA, cH, cV, cD)
# For example, you can reconstruct the image from the coefficients:
# reconstructed_image = pywt.idwt2(coeffs, 'haar')

print("Approximation Coefficients (cA):", cA.shape)
print("Horizontal Detail Coefficients (cH):", cH.shape)
print("Vertical Detail Coefficients (cV):", cV.shape)
print("Diagonal Detail Coefficients (cD):", cD.shape)
```

```python
# prompt: Plot approximation coefficients (LL) , horizontal detail coefficients(LH), vertical
# detail coefficients (HL) , diagonal detail coefficients (HH) results.

import matplotlib.pyplot as plt

# Assuming you have already performed the DWT as shown in the preceding code

# Plot the approximation coefficients (LL)
plt.figure(figsize=(8, 6))
plt.imshow(cA, cmap='gray')
plt.title('Approximation Coefficients (LL)')
plt.show()

# Plot the horizontal detail coefficients (LH)
plt.figure(figsize=(8, 6))
plt.imshow(cH, cmap='gray')
plt.title('Horizontal Detail Coefficients (LH)')
plt.show()

# Plot the vertical detail coefficients (HL)
plt.figure(figsize=(8, 6))
plt.imshow(cV, cmap='gray')
plt.title('Vertical Detail Coefficients (HL)')
plt.show()

# Plot the diagonal detail coefficients (HH)
plt.figure(figsize=(8, 6))
plt.imshow(cD, cmap='gray')
plt.title('Diagonal Detail Coefficients (HH)')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pywt


# Perform 2D Discrete Wavelet Transform (DWT) using the Haar wavelet
coeffs = pywt.dwt2(image_array, 'haar')


# coeffs will contain the approximation and detail coefficients
# (cA, (cH, cV, cD))
cA, (cH, cV, cD) = coeffs


# Calculate the mean across the third dimension for each coefficient array
cA = np.mean(cA, axis=2)
cH = np.mean(cH, axis=2)
cV = np.mean(cV, axis=2)
cD = np.mean(cD, axis=2)


# Display the approximation coefficients (LL):
plt.imshow(cA, cmap='gray')
plt.title('Approximation Coefficients (LL)')
plt.show()


# Display the horizontal detail coefficients (LH):
plt.imshow(cH, cmap='gray')
plt.title('Horizontal Detail Coefficients (LH)')
plt.show()


# Display the vertical detail coefficients (HL):
plt.imshow(cV, cmap='gray')
plt.title('Vertical Detail Coefficients (HL)')
plt.show()


# Display the diagonal detail coefficients (HH):
plt.imshow(cD, cmap='gray')
plt.title('Diagonal Detail Coefficients (HH)')
plt.show()
```
