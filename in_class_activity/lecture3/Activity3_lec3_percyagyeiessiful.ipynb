```python
asphalt_unit_cost = 0
asphalt_quantity = 0
concrete_unit_cost = 0
concrete_quantity = 0
labor_unit_cost = 0
labor_hours = 0
```

```python
asphalt_unit_cost = input("Please enter your asphalt unit cost: ")
asphalt_quantity = input("Please enter your asphalt quantity: ")
concrete_unit_cost = input("Please enter your concrete unit cost: ")
concrete_quantity = input("Please enter your concrete quantity: ")
labor_unit_cost = input("Please enter your labor unit cost: ")
labor_hours = input("Please enter your labor 11hours: ")
```

```python
asphalt_total_cost = int(asphalt_unit_cost)*int(asphalt_quantity)
concrete_total_cost = int(concrete_unit_cost)*int(concrete_quantity)
labor_total_cost = int(labor_unit_cost)*int(labor_hours)
```

```python
overall_project_cost = asphalt_total_cost + concrete_total_cost + labor_total_cost
```

```python
print(asphalt_total_cost)
print(concrete_total_cost)
print(labor_total_cost)
print(overall_project_cost)
```

```python
if asphalt_total_cost < 50000:
  print("under budget")
else:
  print("over budget")
```
