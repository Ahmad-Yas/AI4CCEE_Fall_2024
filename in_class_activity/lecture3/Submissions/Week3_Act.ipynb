```python
# Function to convert HHMMSS int to hours
def convert_to_hours(hhmmss):
    # Convert to string and pad with zeros if necessary
    hhmmss_str = str(hhmmss).zfill(6)

    # Now parse the hours, minutes, and seconds
    hh = int(hhmmss_str[:2])
    mm = int(hhmmss_str[2:4])
    ss = int(hhmmss_str[4:6])

    # Calculate total seconds and convert to hours
    total_seconds = hh * 3600 + mm * 60 + ss
    return total_seconds / 3600  # Convert seconds to hours

# Apply the function to the 'time' column
traffic_data['hours'] = traffic_data['time'].apply(convert_to_hours)

# Display the DataFrame with the new 'hours' column
print(traffic_data)

# Inspect hours column
traffic_data[['hours']].head()
```

```python
# Convert date to datetime format
traffic_data['date'] = pd.to_datetime(traffic_data['date'], format='%Y%m%d')

# Inspect date column
traffic_data[['date']].head()
```

```python
# Get unique dates
unique_dates = traffic_data['date'].unique()

# Loop through each unique date
for date in unique_dates:
    # Filter the DataFrame for the current date
    date_data = traffic_data[traffic_data['date'] == date]

    # Calculate the average lane occupancy
    avg_lane_occupancy = date_data['lane-occupancy'].mean()

    # Print the result
    print(f"Date: {date.date()}, Average Lane Occupancy: {avg_lane_occupancy:.2f}")
```

```python
traffic_data['hours_2'] = traffic_data['hours'].astype(int)

# Inspect date column
traffic_data[['hours_2']].head()
```

```python
# Get unique hours
unique_hours = traffic_data['hours_2'].unique()

# Loop through each unique hour
for hour in unique_hours:
    # Filter the DataFrame for the current hour
    hour_data = traffic_data[traffic_data['hours_2'] == hour]

    # Calculate the average lane occupancy
    avg_lane_occupancy = hour_data['lane-occupancy'].mean()

    # Print the result
    print(f"Hour: {hour}, Average Lane Occupancy: {avg_lane_occupancy:.2f}")
```

```python
# Initialize an empty list to store the results
results = []

# Get unique hours
unique_hours = traffic_data['hours_2'].unique()

# Loop through each unique hour
for hour in unique_hours:
    # Filter the DataFrame for the current hour
    hour_data = traffic_data[traffic_data['hours_2'] == hour]

    # Calculate the average lane occupancy
    avg_lane_occupancy = hour_data['lane-occupancy'].mean()

    # Append the result (hour and average lane occupancy) to the list
    results.append((hour, avg_lane_occupancy))

# Sort the results by average lane occupancy in descending order
sorted_results = sorted(results, key=lambda x: x[1], reverse=True)

# Print the sorted results
for hour, avg_lane_occupancy in sorted_results:
    print(f"Hour: {hour}, Average Lane Occupancy: {avg_lane_occupancy:.2f}")
```

```python
import numpy as np
from PIL import Image, ImageOps, ImageEnhance
```

```python
# Load the image using Pillow
image = Image.open('2018-06-30-18-50.jpg')  # Replace with your image path

# Convert the image to a NumPy array
image_array = np.array(image)

# Display the shape of the NumPy array (this shows dimensions and channels)
print("Image array shape:", image_array.shape)
```

```python
# Load the image using Pillow
image = Image.open('2018-06-30-18-50.jpg')  # image path

# Display the image using Matplotlib
plt.imshow(image)
plt.axis('off')  # Turn off the axis labels
plt.title('Original Image')  # Add a title if you want
plt.show()
```

```python
# Convert the image to grayscale
gray_image = ImageOps.grayscale(image)

# Flip the image horizontally
flipped_image = ImageOps.mirror(gray_image)

# Rotate the image by 90 degrees
rotated_image = gray_image.rotate(90)

# Display the original, grayscale, flipped, and rotated images
plt.figure(figsize=(10, 8))

# Display original image
plt.subplot(2, 2, 1)
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')

# Display grayscale image
plt.subplot(2, 2, 2)
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

# Display flipped image
plt.subplot(2, 2, 3)
plt.imshow(flipped_image, cmap='gray')
plt.title('Flipped Image (Horizontal)')
plt.axis('off')

# Display rotated image
plt.subplot(2, 2, 4)
plt.imshow(rotated_image, cmap='gray')
plt.title('Rotated Image (90 Degrees)')
plt.axis('off')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Convert the image to grayscale (optional step, you can skip if not required)
gray_image = ImageOps.grayscale(image)

# Adjust the brightness
enhancer_brightness = ImageEnhance.Brightness(gray_image)
brighter_image = enhancer_brightness.enhance(1.5)  # Increase brightness by 50%

# Adjust the contrast
enhancer_contrast = ImageEnhance.Contrast(gray_image)
high_contrast_image = enhancer_contrast.enhance(1.5)  # Increase contrast by 50%

# Display the original, brightness-adjusted, and contrast-adjusted images
plt.figure(figsize=(12, 6))

# Display grayscale image
plt.subplot(1, 3, 1)
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

# Display brightness-adjusted image
plt.subplot(1, 3, 2)
plt.imshow(brighter_image, cmap='gray')
plt.title('Increased Brightness')
plt.axis('off')

# Display contrast-adjusted image
plt.subplot(1, 3, 3)
plt.imshow(high_contrast_image, cmap='gray')
plt.title('Increased Contrast')
plt.axis('off')

# Show the plot
plt.tight_layout()
plt.show()
```
