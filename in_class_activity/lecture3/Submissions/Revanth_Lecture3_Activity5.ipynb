Step 5: Convert the image to gray scale. Apply basic transformations such as flipping and rotating.

```python
# Step 1: Import necessary libraries
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Step 2: Open the image (ensure the image path or download process is correct as shown earlier)
img = Image.open('downloaded_image.jpg')

# Step 3: Convert the image to grayscale
gray_img = img.convert('L')

# Step 4: Apply transformations
# Flip the image horizontally
flipped_img = gray_img.transpose(Image.FLIP_LEFT_RIGHT)

# Rotate the image by 90 degrees
rotated_img = gray_img.rotate(90, expand=True)

# Step 5: Plot the original, grayscale, flipped, and rotated images
fig, axes = plt.subplots(1, 4, figsize=(20, 5))

# Original Image
axes[0].imshow(img)
axes[0].set_title("Original Image")
axes[0].axis('off')

# Grayscale Image
axes[1].imshow(gray_img, cmap='gray')
axes[1].set_title("Grayscale Image")
axes[1].axis('off')

# Flipped Image
axes[2].imshow(flipped_img, cmap='gray')
axes[2].set_title("Flipped Image")
axes[2].axis('off')

# Rotated Image
axes[3].imshow(rotated_img, cmap='gray')
axes[3].set_title("Rotated Image")
axes[3].axis('off')

plt.show()

# Step 6: Convert the grayscale image to a NumPy array (optional)
gray_img_array = np.array(gray_img)
print(f"Grayscale Image Array Shape: {gray_img_array.shape}")
```

Step 6: Adjust the brightness and contrast of the image.

```python
# Step 1: Import necessary libraries
from PIL import Image, ImageEnhance
import matplotlib.pyplot as plt

# Step 2: Open the image and convert to grayscale
img = Image.open('downloaded_image.jpg').convert('L')

# Step 3: Adjust Brightness
enhancer_brightness = ImageEnhance.Brightness(img)
bright_img = enhancer_brightness.enhance(1.5)  # Increase brightness (1.0 is original)

# Step 4: Adjust Contrast
enhancer_contrast = ImageEnhance.Contrast(img)
contrast_img = enhancer_contrast.enhance(2.0)  # Increase contrast (1.0 is original)
```

Step 7: Save the manipulated images and Display the manipulated images using Matplotlib.

```python
# Step 5: Save the manipulated images
bright_img.save('brightened_image.jpg')
contrast_img.save('contrast_image.jpg')

print("Images saved successfully!")

# Step 6: Display the original and manipulated images using Matplotlib
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Original Grayscale Image
axes[0].imshow(img, cmap='gray')
axes[0].set_title("Original Grayscale Image")
axes[0].axis('off')

# Brightened Image
axes[1].imshow(bright_img, cmap='gray')
axes[1].set_title("Brightened Image")
axes[1].axis('off')

# Contrast Enhanced Image
axes[2].imshow(contrast_img, cmap='gray')
axes[2].set_title("Contrast Enhanced Image")
axes[2].axis('off')

plt.show()
```
