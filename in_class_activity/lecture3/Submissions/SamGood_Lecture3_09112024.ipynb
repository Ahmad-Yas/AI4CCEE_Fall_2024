```python
# Define variables to store costs and quantities
asphalt_unit_cost = float(input("Enter the unit cost of asphalt (per ton): "))
asphalt_quantity = float(input("Enter the quantity of asphalt (in tons): "))

concrete_unit_cost = float(input("Enter the unit cost of concrete (per cubic yard): "))
concrete_quantity = float(input("Enter the quantity of concrete (in cubic yards): "))

labor_unit_cost = float(input("Enter the labor cost per hour: "))
labor_hours = float(input("Enter the number of labor hours: "))

# Calculate total costs
asphalt_total_cost = asphalt_unit_cost * asphalt_quantity
concrete_total_cost = concrete_unit_cost * concrete_quantity
labor_total_cost = labor_unit_cost * labor_hours

# Compute overall project cost
total_project_cost = asphalt_total_cost + concrete_total_cost + labor_total_cost

# Define the budget
budget = 50000

# Print the costs for each category and the total project cost
print("\nCosts Breakdown:")
print(f"Asphalt Total Cost: ${asphalt_total_cost:.2f}")
print(f"Concrete Total Cost: ${concrete_total_cost:.2f}")
print(f"Labor Total Cost: ${labor_total_cost:.2f}")
print(f"\nTotal Project Cost: ${total_project_cost:.2f}")

# Compare the total project cost to the specified budget
if total_project_cost <= budget:
    print(f"\nThe project is under budget by ${budget - total_project_cost:.2f}.")
else:
    print(f"\nThe project is over budget by ${total_project_cost - budget:.2f}.")
```

```python
import pandas as pd

# Provide the full path to the CSV file in your Google Drive
file_path = '/content/drive/MyDrive/235-Nov.csv'

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(file_path, dtype={'time': str})

# Display the first few rows of the DataFrame to confirm it loaded correctly
print(df.head())

```

```python
# Check the data types of each column
print("Data Types:")
print(df.dtypes)

# Check for any missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Identify numerical columns
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns
print("Numerical Columns:")
print(numerical_columns)

# Get summary statistics (mean, median, std) for numerical columns
numerical_data = df[numerical_columns]

print("\nSummary Statistics for Numerical Columns:")
print(numerical_data.describe())  # This will give count, mean, std, min, 25%, 50% (median), 75%, and max

# Calculate and print median for numerical columns
print("\nMedian Values for Numerical Columns:")
print(numerical_data.median())

# Calculate and print standard deviation for numerical columns
print("\nStandard Deviation for Numerical Columns:")
print(numerical_data.std())

```

```python
df['time'] = df['time'].astype(str)

# Extract the hour from the 'time' column
df['hour'] = df['time'].str[:2].astype(int)
```

```python
average_occupancy_per_day = df.groupby('date')['lane-occupancy'].mean().reset_index()

# Step 3: Display the result
print("Average Lane Occupancy for Each Day:")
print(average_occupancy_per_day)
```

```python
# Calculate average lane occupancy per hour
average_occupancy_per_hour = df.groupby('hour')['lane-occupancy'].mean().reset_index()

# Step 3: Find the hour with the highest average lane occupancy
max_occupancy_row = average_occupancy_per_hour.loc[average_occupancy_per_hour['lane-occupancy'].idxmax()]

# Extract the hour and the maximum average lane occupancy
max_hour = max_occupancy_row['hour']
max_occupancy = max_occupancy_row['lane-occupancy']

# Print the results
print(f"The hour of the day with the highest average lane occupancy is {max_hour}:00 with an occupancy of {max_occupancy:.2f}.")
```

```python
status_counts = df.groupby('status').size().reset_index(name='total_lane_counts')

# Display the result
print("Total Lane Counts by Status:")
print(status_counts)
```

```python
from PIL import Image
image = Image.open('/content/drive/MyDrive/2018-06-30-19-05.jpg')
display(image)
```

```python
import numpy as np
image_array = np.array(image)
```

```python
import numpy as np
import pywt
import matplotlib.pyplot as plt

# Check the shape of the image array
print("Image array shape:", image_array.shape)

# If the image is color (3D array), convert it to grayscale
if len(image_array.shape) == 3 and image_array.shape[2] == 3:
    from skimage.color import rgb2gray
    image_array = rgb2gray(image_array)

# Ensure the image is 2D
if len(image_array.shape) != 2:
    raise ValueError("The image array should be 2D after preprocessing")

# Perform 2D Discrete Wavelet Transform using Haar wavelet
coeffs2 = pywt.dwt2(image_array, 'haar')

# Unpack the coefficients
LL, (LH, HL, HH) = coeffs2

# Plot the results
fig, axes = plt.subplots(2, 2, figsize=(12, 12))
axes[0, 0].imshow(LL, cmap='gray')
axes[0, 0].set_title('Approximation Coefficients (LL)')
axes[0, 0].axis('off')

axes[0, 1].imshow(LH, cmap='gray')
axes[0, 1].set_title('Horizontal Detail Coefficients (LH)')
axes[0, 1].axis('off')

axes[1, 0].imshow(HL, cmap='gray')
axes[1, 0].set_title('Vertical Detail Coefficients (HL)')
axes[1, 0].axis('off')

axes[1, 1].imshow(HH, cmap='gray')
axes[1, 1].set_title('Diagonal Detail Coefficients (HH)')
axes[1, 1].axis('off')

plt.tight_layout()
plt.show()
```

```python
#same data as activity 5 so skipping the load dataset and time preprocessing
df['date'] = pd.to_datetime(df['date'])

unique_dates = df['date'].unique()

# Calculate and print average lane occupancy for each unique date
print("Average Lane Occupancy for Each Unique Date:")
for date in unique_dates:
    average_occupancy = df[df['date'] == date]['lane-occupancy'].mean()
    print(f"{date.date()}: {average_occupancy:.2f}")
```

```python
import matplotlib.pyplot as plt

# Summarize the total lane counts based on different statuses
status_counts = df.groupby('status').size().reset_index(name='total_lane_counts')

# Create a pie chart
plt.figure(figsize=(8, 6))
plt.pie(status_counts['total_lane_counts'], labels=status_counts['status'], autopct='%1.1f%%', startangle=140)
plt.title('Total Lane Counts Based on Status')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is a circle.
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['lane-occupancy'], df['lane-count'], alpha=0.5)  # Adjust alpha for transparency
plt.title('Lane Occupancy vs. Lane Count')
plt.xlabel('Lane Occupancy')
plt.ylabel('Lane Count')
plt.grid()
plt.show()
```
