```python
df = pd.read_csv('/content/235-Nov.csv')
def hhmmss_to_hours(hhmmss):
  hhmmss = f"{int(hhmmss):06d}"
  hh = int(hhmmss[:2])
  mm = int(hhmmss[2:4])
  ss = int(hhmmss[4:])
  return hh + mm/60 + ss/3600
  df['time_column']=df['time'].apply(hhmmss_to_hours)
print(df.head())
```

```python
df = pd.read_csv('/content/235-Nov.csv')
df['date_column']=pd.to_datetime(df['date'],format='%Y%m%d')
unique_dates=df['date_column'].dt.date.unique()
for date in unique_dates:
  rows_for_date=df[df['date_column'].dt.date==date]
  avg_lane_occupancy=rows_for_date['lane-occupancy'].mean()
  print(f"Average lane occupancy for {date}: {avg_lane_occupancy}")
  print(df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/235-Nov.csv')
status_counts = df.groupby('status')['lane-count'].sum()
plt.figure(figsize=(8,8))
plt.pie(status_counts,labels=status_counts.index,autopct='%1.1f%%',startangle=90)
plt.title('Total Lane Counts Based on Status')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/235-Nov.csv')
plt.figure(figsize=(10,6))
plt.scatter(df['lane-count'],df['lane-occupancy'],alpha=0.5)
plt.xlabel('Lane Count')
plt.ylabel('Lane Occupancy')
plt.title('Lane Count vs Lane Occupancy')
plt.grid(True)
plt.show()
```
