```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the data (replace 'data.csv' with your actual file path)
df = pd.read_csv('/content/Wavetronix.csv')

# Convert 'Time' column to datetime format
df['Time'] = pd.to_datetime(df['Time'])

# Fill missing values with the column mean for 'WAVE-NE-Speed'
df['WAVE-NE-Speed'] = df['WAVE-NE-Speed'].fillna(df['WAVE-NE-Speed'].mean())

# Filter the data for a specific day (e.g., day 1)
specific_day = 1
df_day = df[df['day'] == specific_day]

# Apply FFT on the 'WAVE-NE-Speed' column
# First, remove the mean from the data (detrending) to focus on frequency components
speed_data = df_day['WAVE-NE-Speed'] - df_day['WAVE-NE-Speed'].mean()

# Perform FFT
fft_result = np.fft.fft(speed_data)

# Get the corresponding frequencies
fft_freq = np.fft.fftfreq(len(speed_data), d=1)  # assuming data is sampled every minute (d=1 minute)

# Plot the magnitude of the FFT result
plt.figure(figsize=(10, 6))

# Plot the absolute value of the FFT (magnitude spectrum)
plt.plot(fft_freq[:len(fft_freq)//2], np.abs(fft_result)[:len(fft_result)//2], color='blue')

# Labels and title
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.title(f'FFT of WAVE-NE-Speed on Day {specific_day}')

# Show grid
plt.grid(True)

# Show plot
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# Load the data (replace 'data.csv' with your actual file path)
df = pd.read_csv('/content/Wavetronix.csv')

# Convert 'Time' column to datetime format
df['Time'] = pd.to_datetime(df['Time'])

# Fill missing values with the column mean for 'WAVE-NE-Speed'
df['WAVE-NE-Speed'] = df['WAVE-NE-Speed'].fillna(df['WAVE-NE-Speed'].mean())

# Filter the data for a specific day (e.g., day 1)
specific_day = 1
df_day = df[df['day'] == specific_day]

# Extract WAVE-NE-Speed data
speed_data = df_day['WAVE-NE-Speed']

# Find peaks in the WAVE-NE-Speed data
peaks, _ = find_peaks(speed_data)

# Calculate the average peak speed
average_peak_speed = np.mean(speed_data.iloc[peaks])

# Print the average peak speed
print(f"Average Peak Speed for Day {specific_day}: {average_peak_speed:.2f} km/h")

# Plot the original speed data and mark the peaks
plt.figure(figsize=(10, 6))
plt.plot(df_day['Time'], speed_data, label='WAVE-NE-Speed', color='blue', alpha=0.5)
plt.plot(df_day['Time'].iloc[peaks], speed_data.iloc[peaks], 'ro', label='Peaks')  # Mark peaks

# Labels and title
plt.xlabel('Time')
plt.ylabel('Speed (km/h)')
plt.title(f'WAVE-NE-Speed with Peaks on Day {specific_day}')

# Show legend and grid
plt.legend()
plt.grid(True)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Adjust layout to prevent overlap
plt.tight_layout()

# Show plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pywt
from PIL import Image

# Load the image (replace 'image4.jpg' with your actual image path)
image_path = '/content/2018-06-30-19-05.jpg'
image = Image.open(image_path).convert('L')  # Convert to grayscale
plt.imshow(image, cmap='gray')
plt.axis('off')  # Hide axes
plt.title('Original Image')
plt.show()

# Convert image to NumPy array
image_array = np.array(image)
print("Image shape:", image_array.shape)

# Perform wavelet transform using Haar wavelet
coeffs = pywt.dwt2(image_array, 'haar')

# Unpack the coefficients
LL, (LH, HL, HH) = coeffs

# Plot the results
plt.figure(figsize=(12, 12))

plt.subplot(2, 2, 1)
plt.imshow(LL, cmap='gray')
plt.title('Approximation Coefficients (LL)')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(LH, cmap='gray')
plt.title('Horizontal Detail Coefficients (LH)')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(HL, cmap='gray')
plt.title('Vertical Detail Coefficients (HL)')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(HH, cmap='gray')
plt.title('Diagonal Detail Coefficients (HH)')
plt.axis('off')

plt.tight_layout()
plt.show()
```
