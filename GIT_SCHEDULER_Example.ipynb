```python
import requests

# Set up your credentials and repository information
GITHUB_TOKEN = 'YOUR_TOKEN_FROM_HERE'  # LINK  'https://github.com/settings/tokens' Replace with your actual token

# EXAMPLE WITH MINE

REPO_OWNER = 'miguel-source'
REPO_NAME = 'AI4CCEE_Fall_2024'
BRANCH_NAME = 'main'

headers = {
    'Authorization': f'token {GITHUB_TOKEN}',
    'Accept': 'application/vnd.github.v3+json'
}

# List of files to delete (use relative paths from the repository's root)
files_to_delete = [
    'letcture3.ipynb',
    'activity4.ipynb',
    'activity2.ipynb',
    'Miguel_Diaz_Activity_5.ipynb',
    'Miguel_Diaz_Act4.ipynb',
    'Miguel_Diaz_Act3.ipynb',
    'HW6_Miguel_Diaz.ipynb',
    'Class4.ipynb'
]

for file_path in files_to_delete:
    # Get the file's SHA
    url = f'https://api.github.com/repos/{REPO_OWNER}/{REPO_NAME}/contents/{file_path}'
    response = requests.get(url, headers=headers)

    # Handle potential errors with fetching the file data
    if response.status_code != 200:
        print(f"Failed to fetch file metadata for {file_path}: {response.json()}")
        continue

    file_data = response.json()
    file_sha = file_data['sha']

    # Delete the file
    delete_url = f'https://api.github.com/repos/{REPO_OWNER}/{REPO_NAME}/contents/{file_path}'
    data = {
        'message': f'Delete {file_path}',
        'sha': file_sha,
        'branch': BRANCH_NAME
    }
    delete_response = requests.delete(delete_url, headers=headers, json=data)

    if delete_response.status_code == 200:
        print(f'Successfully deleted {file_path}')
    else:
        print(f'Failed to delete {file_path}: {delete_response.json()}')

```

```python
import requests
import base64

# Step 1: Set up your GitHub credentials and repository information
GITHUB_TOKEN = 'YOUR_TOKEN_FROM_HERE'  # LINK  'https://github.com/settings/tokens' Replace with your actual token
REPO_OWNER = 'miguel-source'
REPO_NAME = 'AI4CCEE_Fall_2024'
BRANCH_NAME = 'main'
FILE_PATH = 'in_class_activity/lecture4/Submissions/Class4_Miguel_Diaz.ipynb'  # Target path in GitHub
LOCAL_FILE_PATH = '/content/drive/MyDrive/Colab Notebooks/Class4_Miguel_Diaz.ipynb'  # Path in Google Drive

# Step 2: Read and encode the file from Google Drive
with open(LOCAL_FILE_PATH, 'rb') as file:
    file_content = file.read()
    encoded_content = base64.b64encode(file_content).decode('utf-8')

# Step 3: Set up the API URL for creating or updating a file in GitHub
url = f"https://api.github.com/repos/{REPO_OWNER}/{REPO_NAME}/contents/{FILE_PATH}"

# Step 4: Prepare the payload for the request
data = {
    "message": "Add Class5_Miguel_Diaz.ipynb",
    "content": encoded_content,
    "branch": BRANCH_NAME
}

# Step 5: Set up the headers for authentication
headers = {
    'Authorization': f'token {GITHUB_TOKEN}',
    'Accept': 'application/vnd.github.v3+json'
}

# Step 6: Make the PUT request to upload the file to GitHub
response = requests.put(url, headers=headers, json=data)

# Step 7: Check the response
if response.status_code == 201:
    print(f"File {FILE_PATH} successfully uploaded.")
elif response.status_code == 200:
    print(f"File {FILE_PATH} was updated.")
else:
    print(f"Failed to upload file: {response.json()}")
```
